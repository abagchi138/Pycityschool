```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
totalschools= school_data_complete['school_name'].nunique()
#totalschools
totalstudents= school_data_complete['Student ID'].nunique()
#totalstudents
totalbudget= school_data_complete['budget'].unique().sum()
#totalbudget
avgmathscore= school_data_complete['math_score'].mean()
#avgmathscore
avgreadingscore= school_data_complete['reading_score'].mean()
#avgreadingscore
passingmath=round((len(school_data_complete[school_data_complete["math_score"] >= 70])/totalstudents)*100, 6)
#passingmath
passingreading=round((len(school_data_complete[school_data_complete["reading_score"] >= 70])/totalstudents)*100, 6)
#passingreading
bothpassing= ((school_data_complete[(school_data_complete['math_score']>=70)& 
                                     (school_data_complete['reading_score']>=70)].count()["student_name"])/totalstudents)*100
#bothpassing

DistrictSummary_df= pd.DataFrame({
    "total schools": [totalschools],
    "total budget": [totalbudget],
    "Average Math Score": [avgmathscore],
    "Average Reading Score": [avgreadingscore],
    "Total Students": [totalstudents],
    "Passing Math": [passingmath],
    "Passing Reading": [passingreading],
    "Overall Passing": [bothpassing],
})
DistrictSummary_df



```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python

schooltype= school_data.set_index(["school_name"])["type"]
#schooltype

schoolstudentcount= school_data_complete["school_name"].value_counts()
#schoolstudentcount

#per school budget
perschoolbudget= school_data_complete.groupby(["school_name"]).mean()["budget"]

#per school capita
perschoolcapita= perschoolbudget/ schoolstudentcount

#perschoolcapita

#avg scores by school
avg_math = school_data_complete.groupby(["school_name"]).mean()["math_score"]
#avg_math

avg_reading = school_data_complete.groupby(["school_name"]).mean()["reading_score"]
#avg_reading

#passing math
passing_math= school_data_complete[school_data_complete['math_score']>= 70]
#passing_math

#passing reading
passing_reading= school_data_complete[school_data_complete['reading_score']>= 70]
#passing_reading

#overall passing

overall_passing= school_data_complete[(school_data_complete['math_score']>= 70) &
                                       (school_data_complete['reading_score']>= 70)]
#overall_passing

#% percent passing math
percentpassingmath= passing_math.groupby(["school_name"]).count()["student_name"]/schoolstudentcount*100
#percentpassingmath

#% percent passing reading
percentpassingreading= passing_reading.groupby(["school_name"]).count()["student_name"]/schoolstudentcount*100
#percentpassingreading

#% percent passing overall
overallpassingpercentage= overall_passing.groupby(["school_name"]).count()["student_name"]/schoolstudentcount*100 
#overallpassingpercentage

Schoolsummary_df= pd.DataFrame({
    "School Type": schooltype,
    "Total Students": schoolstudentcount,
    "Total Student Budget": perschoolbudget,
    "Per Student Budget": perschoolcapita,
    "Average Math scores": avg_math,
    "Average Reading Scores": avg_reading,
    "% passing math": percentpassingmath,
    "% passing reading": percentpassingreading,
    "% total students passing": overallpassingpercentage
    
})

Schoolsummary_df

```
